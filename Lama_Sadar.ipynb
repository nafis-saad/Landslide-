```python
import seaborn as sns

plt.figure(figsize=(16,9))
ax = sns.lineplot(x="SL", y="2010", marker = "o", palette = "husl",data=df)
ax = sns.lineplot(x="SL", y="2015", marker = "o", palette = "husl",data=df)
ax = sns.lineplot(x="SL", y="2016", marker = "o", palette = "husl",data=df)
ax = sns.lineplot(x="SL", y="2017", marker = "o", palette = "husl",data=df)
ax = sns.lineplot(x="SL", y="2018", marker = "o", palette = "husl",data=df)


ax.set_xlabel("Days", fontsize = 10)
ax.set_ylabel("Rainfall", fontsize = 12)
plt.legend(["2010", "2015", "2016","2017", "2018"])

import matplotlib.pyplot as plt

plt.axhline(y=50, color = 'r', linestyle = 'dotted', label = " 50 mm Rainfall")
plt.axhline(y=100, color = 'b', linestyle = 'dotted', label = " 100 mm Rainfall")
plt.axhline(y=150, color = 'g', linestyle = 'dotted', label = " 100 mm Rainfall")
ax.axis([0, 100, 0, 350])
ax.set_title("Landlslide Events in Lama Upazila Since 2010",fontsize = 18)

x_ticks = list(range(1,95,2))
x_labels = x_ticks
plt.xticks(ticks=x_ticks, labels=x_labels)

plt.savefig('/content/drive/MyDrive/Python_Practice.png')
plt.show() 


```
